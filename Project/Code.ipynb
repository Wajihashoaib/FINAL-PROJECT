# Cleaning of Data

```python
import pandas as pd
import numpy as np
import csv
```

```python
data = pd.read_csv('Total_reviews.csv')
```

```python
data.head()
```

```python
data.shape
```

# Drop Unnecessary Columns

```python
data.drop(columns=data.columns[:3], axis=1, inplace=True)
```

```python
data.head()
```

# Creating Column Category from sourcefilename

```python
data['Category'] = data['sourcefilename'].replace('review_','', regex=True)
```

```python
data.head()
```

```python
data['Category'] = data['Category'].replace('.csv','', regex=True)
```

```python
data.head()
```

# Drop Column SourceFileName

```python
data.drop(columns=data.columns[-2], axis=1, inplace=True)
```

```python
data.head()
```

# Remove dollar and Unnecessary Characters 

```python
data['Price'] = [x.strip('$') for x in data.Price]
data['Price'] = [x.strip('&') for x in data.Price]
```

```python
data.head()
```

# Extracting Year from Date

```python
data['Date'] = data['Date'].str[-4:]
data.head()
```

# Extracting Ratings 

```python
data['Ratings'] = data['Ratings'].str[:1]
```

```python
data.head()
```

```python
data.to_csv('Final_Dataset.csv')
```
